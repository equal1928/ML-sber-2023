```python
import pandas as pd
```

```python
data = pd.read_pickle('./data_entities.pickle')
```

```python
texts = data['text']
labels = data['tag']
```

```python
ent_dict = {}
for text, tags in zip(texts, labels):
    for word, tag in zip(text, tags):
        if tag != "O":
            if tag not in ent_dict.keys():
                ent_dict[tag] = []
```

```python
previous_tag = 'O'
seq_words = []

for text, tags in zip(texts, labels):
    for word, tag in zip(text, tags):
        if previous_tag == tag:
            if tag != 'O':
                seq_words.append(word)
                previous_tag = tag
            
        elif previous_tag != tag:
            if previous_tag == "O":
                seq_words.append(word)
                previous_tag = tag
                
            else:
                ent_dict[previous_tag].append(' '.join(seq_words))
                seq_words = []
                previous_tag = tag
                
```

```python
for key in ent_dict:
    ent_dict[key] = list(set(ent_dict[key]))
```

```python
for key in ent_dict:
    print(key, len(ent_dict[key]))
```
